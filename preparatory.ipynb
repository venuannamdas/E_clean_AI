Title: The clean Communication - The Agenda of tomorrow for Managers, Engineers and Doctors, towards a cleaner Email Visualization using smart AI algorithms

Goal: The response recommendations would be Reply, Delete (time_waste) Save or
Later. Proposed solution is to build a email ranker which lists
a priority based email set, this would be augmented with a
recommendation system which provides a probable course of
action which would be performed on the emails

this part of the code involves preparation of dataset. The Topic of the project is suggested by Dr David Woon, a Director at Continental Automotive Singapore. So the datset is modified to suit his suggestions

The data is enormous and only around 9571 observations are considered for the project.


```python
e_data='https://raw.githubusercontent.com/venuannamdas/E_clean_AI/master/CAPSTONE.csv'
```

```python
import pandas as pd
```

```python
df=pd.read_csv(e_data)
```

```python
df.head
```

```python
df.head()
```


Modify
1. change kay.mann@enron.com to david.woon@smartcontinental.com


```python
# Replace 'kay.mann@enron.com' with the new value you want, e.g., 'new_email@example.com'
new_value = 'david.woon@smartcontinental.com'

# Use the DataFrame's applymap function to replace the old value in all columns
df = df.applymap(lambda x: new_value if x == 'kay.mann@enron.com' else x)
```

```python
# Replace 'kmann' with 'davidw'
df = df.applymap(lambda x: x.replace('kmann', 'davidw') if isinstance(x, str) else x)
```

```python
df.head
```

```python
# Replace 'mann-k' with 'david-w'
df = df.applymap(lambda x: x.replace('mann-k', 'david-w') if isinstance(x, str) else x)
```

```python
# Replace ' kay.mann' with 'woon.david'
df = df.applymap(lambda x: x.replace(' kay.mann', 'woon.david') if isinstance(x, str) else x)
```

```python
# Replace ' kay' with 'woon'
df = df.applymap(lambda x: x.replace(' kay', 'woon') if isinstance(x, str) else x)
```

```python
df = df.applymap(lambda x: x.replace(' mann', 'david') if isinstance(x, str) else x)
```

```python
df.head
```

```python
# Replace 'kmann' with 'davidw'
df = df.applymap(lambda x: x.replace('KMANN', 'davidw') if isinstance(x, str) else x)
```

```python
df.head()
```

```python
df = df.applymap(lambda x: x.replace('delete', 'time_waste') if isinstance(x, str) else x)
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Label'].unique()
```

```python
# label change  "required_response"
df = df.applymap(lambda x: x.replace('thread', 'required_response') if isinstance(x, str) else x)
```

```python
df['Label'].unique()
```

david-w is our user, for whom we are ranking all his emails according to Labels.

(1) time_waste: If the email consisted of
promotional data, out of scope information for the user, spam
mails and of it was an end of the conversation it was labelled
as time_waste.

(2)reply: If the email contains a reply then it was tagged as
reply (sent)

(3)required_response: if the email was not replied and did have
some important information it was tagged as thread
